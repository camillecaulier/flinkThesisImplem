# Running the benchmark
We will run our benchmark multiple times on the remote server, fetch the results, aggregate them, and plot the results.

```python
import os
cur_dir=os.getcwd() # save current directory to save the generated CSV files
```

# Preparing the experiments
We first create a csv containing informations about the current experiment, containing the following important pieces of information: 
- `exp_id`
- `date`
- `notebook_dir`
- `state`

Each row will contain a single benchmark running on the remote server.

```python
import pandas
import io
import uuid
import datetime
import pathlib

notebook_id = str(uuid.uuid4()) # Generate ID for current experiment

experiments = "exp_id\n" # header

for i in range(5):
    experiments += "%d\n" % (i)
    
experiments=io.StringIO(experiments) # Convert to StringIO object

df = pandas.read_csv(experiments) # Convert to Panda DataFrame

df['date'] = str(datetime.date.today())
df['notebook_dir'] = ["%s/experiments/%s/%s" % (cur_dir,notebook_id, i['exp_id']) for _,i in df.iterrows()]
df['state'] = 'created'

# randomize test execution
pathlib.Path("%s/experiments/%s" % (cur_dir, notebook_id)).mkdir(exist_ok=True, parents=True)
df.to_csv("%s/experiments/%s/experiment.csv" % (cur_dir, notebook_id))
df
```

# Running the experiments
We are now ready to run each experiment contained in our csv. Papermill will help us in running a notebook with variables specified at runtime!
For each row in our experiment file, a new directory will be created holidng the notebook that was effectively executed along with the results produced by the notebook.

For example, after the execution of the first row a new directory named `{cur_dir}/experiments/{notebook_id}/0` will be created with the following files:
- `experiment.ipynb`: the notebook
- `output.dat`: the results

⚠️ We expect you to have already pushed the benchmark Jar on the remote server as explained in the `setup.ipynb` notebook.

```python
import papermill as pm
df = pandas.read_csv("%s/experiments/%s/experiment.csv" % (cur_dir, notebook_id))
for index, exp in df.iterrows():
    if exp['state'] != 'created':
        continue

    print("Running experiment number: %d, ID: %s" % (index, exp['exp_id']))
    pathlib.Path(exp['notebook_dir']).mkdir(exist_ok=True, parents=True)
    
    # run notebook
    pm.execute_notebook(
        'single_experiment.ipynb',
        '%s/experiment.ipynb' % (exp['notebook_dir']),
        parameters = dict(notebook_dir = exp['notebook_dir'])
    )
    # change and save new state
    df.at[index, 'state'] = 'executed'
    df.to_csv("%s/experiments/%s/experiment.csv" % (cur_dir, notebook_id))
```

# Retrieving the results
We are now ready to concatenate the multiple experiments into a single result file.
We can then do some custom aggregation and display the results

```python
frames = []

for index, exp in df.iterrows():
    output = pandas.read_csv("%s/output.dat" % (exp['notebook_dir']))
    frames.append(output)
    
results = pandas.concat(frames)
display(results)
# Save results in same dir as 'experiment.csv'
results.to_csv("%s/experiments/%s/benchmark.dat" % (cur_dir, notebook_id)) 

# Aggregate results and plot
results.groupby(['Lock', 'Threads'], as_index=False)['Duration'].mean().set_index('Threads').groupby(['Lock'])['Duration'].plot(legend=True)
```
