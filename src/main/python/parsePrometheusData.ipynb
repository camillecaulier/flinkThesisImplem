```python

cur_dir=os.getcwd() # save current directory to save the generated CSV files
experimentName ="100000_150_javaSource_p_6"
experimentDirectory = "%s/experiments/experiment_%s/" % (cur_dir,experimentName)
mainDFLocation =experimentDirectory+ "dataframe.csv"
# pathlib.Path(experimentDirectory).mkdir(exist_ok=True, parents=True)
n_experiments = 5
main_parallelism = 6
```

```python

df = pd.DataFrame()

def checkLineIsValid(line):
    return "job name" in line

def parse_and_correct_json_line(line):
    # Replace single quotes with double quotes and ensure the string is a valid JSON format
    corrected_line = line.replace("'", '"')
    return corrected_line

def extract_name_and_data(input_string):
    pattern = r"Benchmarking operator:\s*(\w+)\s*with file:\s*(\w+),(\d+),(\d+),(\d+),([\d.]+)"
    match = re.search(pattern, input_string)

    if match:
        name = match.group(1)
        data = match.group(2), int(match.group(3)), int(match.group(4)), int(match.group(5)), float(match.group(6))
        result = [name] + list(data)
        return result
    else:
        # Return an empty list if the pattern doesn't match
        return []

for i in range(n_experiments):
    #read all experiments
    file_path = experimentDirectory +str(i)+"/"+"throughput.dat"
    with open(file_path, 'r') as file:
        lines = file.readlines()
        for line in lines:
            if checkLineIsValid(line):
                # line = parse_and_correct_json_line(line)
                # instanceData = json.loads(line)
                instanceData = eval(line)
                print(instanceData['job name'])
```
